# Rank based filters

## Rank vs. weighted sum

Compared to what was seen in the former section, if the local processing is different from a weighted sum of neighbour pixels, the said local processing is named a non-linear filter. An important category of those non-linear filters are the rank filters. These filters use the ranked levels from the neighbourhood.
|Illustration of a rank filter, here the median filter|
|:-:|
|<img src="../data/median.png" alt="Illustration of a rank filter, here the median filter" title="Illustration of a rank filter, here the median filter" width="300"/>

An illustration of such a rank filter, the median filter, is given in the image above, and a concrete example is given in the code below.

```python
from skimage.data import camera
import matplotlib.pyplot as plt
import numpy as np
from skimage.filters import rank as skr
from skimage.morphology import disk

# Original image
ima = camera()

# Salt and pepper noise
n = np.random.random(ima.shape)
noised_ima = ima.copy()
noised_ima[n<.05] = 0
noised_ima[n>.95] = 255

# Median filter
filtered_im = skr.median(noised_ima,disk(1))

plt.figure(figsize=[20,10])
plt.subplot(1,3,1)
plt.title('Original image')
plt.imshow(ima,cmap=plt.cm.gray)
plt.subplot(1,3,2)
plt.title('Noised image')
plt.imshow(noised_ima,cmap=plt.cm.gray);
plt.subplot(1,3,3)
plt.title('Filtered image')
plt.imshow(filtered_im,cmap=plt.cm.gray);
```

The size of the neighbourhood can be changed, it can be widened : which leads to a better noise filtering but also to a more blurry image, like shown in the example below. The profile of the filtered image, compared to the profile of the original image, clearly shows the nature of the filter. The red line is the followed scan line of the profile.
> A profile is different from a histogram : a profile represents the intensity of an image measured along a scan line. The intensity corresponds to the value of the pixel in the HSV space.

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from scipy import ndimage

from skimage.data import camera
import skimage.filters.rank as skr
from skimage.morphology import disk

def profile(ima,p0,p1,num):
    n = np.linspace(p0[0],p1[0],num)
    m = np.linspace(p0[1],p1[1],num)
    return [n,m,ndimage.map_coordinates(ima, [m,n], order=0)]

im = camera()[-1::-2,::2]

#filtered version
radius = 3
rank = skr.median(im,disk(radius))
[x,y,p] = profile(im,(53,200),(73,164),100)
[x,y,prank] = profile(rank,(53,200),(73,164),100)

plt.figure(figsize=[20,10])
plt.subplot(1,2,1)
plt.imshow(rank,interpolation='nearest',cmap=cm.gray,origin='lower')
title = 'Local median (radius = ' + str(radius) + ')'
plt.title(title)
plt.plot([53,73], [200,164], color='red', linewidth=2)
plt.subplot(1,2,2)
plt.plot(p,label='original')
plt.plot(prank,label='median')
plt.title('profile')
plt.legend();
```

Just to illustrate the difference with a mean filter with the same neighbourhood as the former example, the next code applies a mean filter and shows the correspondent histogram. As shown, the histogram is way more average concerning the values, which indicates a less drastic choice in the values of gray.

```python
#filtered version
radius = 3
mean = skr.mean(im,disk(radius))
[x,y,p] = profile(im,(53,200),(73,164),100)
[x,y,pmean] = profile(mean,(53,200),(73,164),100)

plt.figure(figsize=[20,10])
plt.subplot(1,2,1)
plt.imshow(mean,interpolation='nearest',cmap=cm.gray,origin='lower')
title = 'Local mean (radius = ' + str(radius) + ')'
plt.title(title)
plt.plot([53,73], [200,164], color='red', linewidth=2)
plt.subplot(1,2,2)
plt.plot(p,label='original')
plt.plot(pmean,label='mean')
plt.title('profile')
plt.legend();
```

## Local histogram method



The local histogram method is used to equalize an image. It is the same principle as the global histogram equalization. The main difference is that the histogram is computed on a neighbourhood around a certain pixel : the equalization is done locally. The following illustrations explains the formation of the local histograms.
|First histogram|Second histogram, computed after translation of the pixel and its neighbourhood|Path of the computations of the following histograms|
|:-:|:-:|:-:|
|<img src="../data/histo1.png" alt="First histogram" title="First histogram" width="400"/>|<img src="../data/histo2.png" alt="Second histogram, computed after translation of the pixel and its neighbourhood" title="Second histogram, computed after translation of the pixel and its neighbourhood" width="400"/>|<img src="../data/histo3.png" alt="Path of the computations of the following histograms" title="Path of the computations of the following histograms" width="400"/>|

The next code is an example directly taken from https://scikit-image.org/. It clearly shows the difference between a global equalization and a local equalization, the latter being far more precise when it comes to accentuate the contrast for precise elements.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from skimage import data
from skimage.util.dtype import dtype_range
from skimage.util import img_as_ubyte
from skimage import exposure
from skimage.morphology import disk
from skimage.morphology import ball
from skimage.filters import rank

matplotlib.rcParams['font.size'] = 9

def plot_img_and_hist(image, axes, bins=256):
    """Plot an image along with its histogram and cumulative histogram.

    """
    ax_img, ax_hist = axes
    ax_cdf = ax_hist.twinx()

    # Display image
    ax_img.imshow(image, cmap=plt.cm.gray)
    ax_img.set_axis_off()

    # Display histogram
    ax_hist.hist(image.ravel(), bins=bins)
    ax_hist.ticklabel_format(axis='y', style='scientific', scilimits=(0, 0))
    ax_hist.set_xlabel('Pixel intensity')

    xmin, xmax = dtype_range[image.dtype.type]
    ax_hist.set_xlim(xmin, xmax)

    # Display cumulative distribution
    img_cdf, bins = exposure.cumulative_distribution(image, bins)
    ax_cdf.plot(bins, img_cdf, 'r')

    return ax_img, ax_hist, ax_cdf

# Load an example image
img = img_as_ubyte(data.moon())

# Global equalize
img_rescale = exposure.equalize_hist(img)

# Equalization
footprint = disk(30)
img_eq = rank.equalize(img, footprint=footprint)

# Display results
fig = plt.figure(figsize=(20, 10))
axes = np.zeros((2, 3), dtype=object)
axes[0, 0] = plt.subplot(2, 3, 1)
axes[0, 1] = plt.subplot(2, 3, 2, sharex=axes[0, 0], sharey=axes[0, 0])
axes[0, 2] = plt.subplot(2, 3, 3, sharex=axes[0, 0], sharey=axes[0, 0])
axes[1, 0] = plt.subplot(2, 3, 4)
axes[1, 1] = plt.subplot(2, 3, 5)
axes[1, 2] = plt.subplot(2, 3, 6)

ax_img, ax_hist, ax_cdf = plot_img_and_hist(img, axes[:, 0])
ax_img.set_title('Low contrast image')
ax_hist.set_ylabel('Number of pixels')

ax_img, ax_hist, ax_cdf = plot_img_and_hist(img_rescale, axes[:, 1])
ax_img.set_title('Global equalise')

ax_img, ax_hist, ax_cdf = plot_img_and_hist(img_eq, axes[:, 2])
ax_img.set_title('Local equalize')
ax_cdf.set_ylabel('Fraction of total intensity')

# prevent overlap of y-axis labels
fig.tight_layout()
```

## Local maximum and local minimum

Local minimum value and local maximum value are special cases of the ranked filters.
The figure bellow illustrates the effect of respectively replacing the pixel by...
1. the highest value of its neighboorhood, the local maximum, 
2. the lowest value of its neighboorhood, the local minimum.

```python
#filtered version
radius = 5
selem = disk(radius)
rank1 = skr.maximum(im,selem)
rank2 = skr.minimum(im,selem)
rank3 = skr.gradient(im,selem)
[x,y,p] = profile(im,(53,200),(73,164),100)
[x,y,prank1] = profile(rank1,(53,200),(73,164),100)
[x,y,prank2] = profile(rank2,(53,200),(73,164),100)
[x,y,prank3] = profile(rank3,(53,200),(73,164),100)

fig = plt.figure(figsize=[24,9])
plt.subplot(1,3,1)
plt.imshow(rank1,interpolation='nearest',cmap=cm.gray,origin='lower')
plt.title('Local maximum (radius = %d)'%radius)
plt.plot([53,73], [200,164], color='red', linewidth=2)
plt.subplot(1,3,2)
plt.imshow(rank2,interpolation='nearest',cmap=cm.gray,origin='lower')
plt.title('Local minimum (radius = %d)'%radius)
plt.plot([53,73], [200,164], color='red', linewidth=2)
plt.subplot(1,3,3)
plt.plot(p,label='Original')
plt.plot(prank1,label='maximum')
plt.plot(prank2,label='minimum')
plt.title('Profile')
plt.gca().set_ylim([0,210])
plt.legend(loc=5);
```

### Local autolevel

As introduced in the former section, the auto-level LUT remaps the dynamic scale of an image to the full dynamic scale of all possible shades, and by doing so extends the range of values of the image. The function of the autolevel filter is given by the following equation :
$$g_{out} = 255 \cdot \frac{g_{in} - g_{min}}{g_{max}-g_{min}}$$
where $g_{min}$ and $g_{max}$ are respectively the minimimal and maximal values present in the image. In te case of a local autolevel filter, this remapping is done on a neighbourhood scale, and thus recomputed for each pixel.

```python
#local auto-level
from skimage.io import imread
ima = imread('../data/bones.png')
rank = skr.autolevel(ima,disk(5))

def compare(f,g,roi=None):
    plt.figure(figsize=[10,10])
    plt.subplot(1,2,1)
    plt.imshow(f,cmap=plt.cm.gray)
    plt.gca().invert_yaxis()
    plt.subplot(1,2,2)
    plt.imshow(g,cmap=plt.cm.gray)
    plt.gca().invert_yaxis()
    if roi is not None:
        f = f[roi[0]:roi[1],roi[2]:roi[3]]
        g = g[roi[0]:roi[1],roi[2]:roi[3]]
        plt.figure(figsize=[10,10])
        plt.subplot(1,2,1)
        plt.imshow(f,cmap=plt.cm.gray)
        plt.gca().invert_yaxis()
        plt.subplot(1,2,2)
        plt.imshow(g,cmap=plt.cm.gray)
        plt.gca().invert_yaxis()

compare(ima,rank)
```

### Local morphological contrast enhancement

This filter enhances contrast of an image by replacing each pixel by...
- the local maximum if the pixel grayvalue is closer to the local maximum than the local minimum,
- the local minimum if the pixel grayvalue is closer to the local minimum than the local maximum.

Two versions of the filter exist : the "classic" one and the "percentile" one. For the latter, it is just the same as the autolevel with percentile : only the grayvalues between the two percentiles (of the cummulative histogram) given in argument are considered in the filter. An example of a local contrast enhancement filter with percentiles is given in the code below.

```python
# morphological contrast enhancement with percentiles
rank = skr.enhance_contrast_percentile(im,disk(2),p0=.1,p1=.9)
[x,y,p] = profile(im,(53,200),(73,164),100)
[x,y,prank] = profile(rank,(53,200),(73,164),100)

plt.figure(figsize=[20,10])
plt.subplot(1,2,1)
plt.imshow(rank,interpolation='nearest',cmap=cm.gray,origin='lower')
plt.gca().set_xlim((0,255))
plt.gca().set_ylim((0,255))
plt.plot([53,73], [200,164], color='red', linewidth=2)
plt.subplot(1,2,2)
plt.plot(p,label='Original')
plt.plot(prank,label='Local contrast enhancement (with percentiles)')
plt.legend();
```

## Local threshold

This filter uses the concept of threshold introduced before. The function for the threshold LUT is given by the next equation.
$$
\begin{align*}
g_{out} =
\begin{cases}
    255 & \textrm{ if } g_{in}>th \\
    0 & \textrm{ otherwise }
\end{cases}
\end{align*}
$$

In the case of a local threshol filter, the LUT is applied to the neighbourhood of every pixel. The example given by the code below illustrates

```python
p0 = (53,200)
p1 =(73,164)

low = im<=(skr.minimum(im,disk(10))+10)
[x,y,p] = profile(im, p0, p1,100)
[x,y,plow] = profile(low, p0, p1,100)

plt.figure(figsize=[20,10])
plt.subplot(1,2,1)
plt.imshow(low,interpolation='nearest',cmap=cm.gray,origin='lower')
plt.plot([53,73], [200,164], color='red', linewidth=2)
plt.gca().set_xlim((0,255))
plt.gca().set_ylim((0,255))
ax1 = plt.subplot(1,2,2)
ax2 = ax1.twinx()
ax1.plot(p,label='Original')
ax2.plot(plow,'g',label='Lowest')
ax2.legend()

high = im>=(skr.maximum(im,disk(10))-10)
[x,y,p] = profile(im, p0, p1,100)
[x,y,phigh] = profile(high, p0, p1,100)

plt.figure(figsize=[20,10])
plt.subplot(1,2,1)
plt.imshow(high,interpolation='nearest',cmap=cm.gray,origin='lower')
plt.plot([53,73], [200,164], color='red', linewidth=2)
plt.gca().set_xlim((0,255))
plt.gca().set_ylim((0,255))
ax1 = plt.subplot(1,2,2)
ax2 = ax1.twinx()
ax1.plot(p,label='Original')
ax2.plot(phigh,'g',label='Highest')
ax2.legend()
```

## Anisotropic filter

Anisotropic filtering is a method for improving the quality of an image with textures, surfaces that are at oblique viewing angles with respect to the camera, where the projection of said texture or surface appears to be non-orthogonal. In other words, an anisotropic filter does not filter the same way in each direction.

Like bilinear filters, anisotropic filters eliminate aliasing effects, but improve on these other techniques by reducing blur and preserving detail at extreme viewing angles. The anisotropic filtering uses a different kind of mapping in storage, like illustrated below. The space taken in memory is greater, but it results in a better filtering.

### Diffusion filter

Anisotropic diffusion, or more simply diffusion filtering, is a technique that reduces the noise in an image without removing significant parts of the content, typically edges, lines or other details that are important for the interpretation of the image. Diffusion filtering is similar to the process that creates a scale space, where an image generates a parameterized family of successively more and more blurred images based on a diffusion process, generally a 2D isotropic Gaussian filter. Each of the resulting images in this family are given as a convolution between the image and a the filter, where the width of the filter increases with the parameter. This diffusion process is a linear and space-invariant transformation of the original image. It is what was presented before as a mipmap.

Diffusion filtering is a generalization of this diffusion process : it produces a family of parameterized images, but each resulting image is a combination between the original image and a filter that depends on the local content of the original image. In consequence, a diffusion filter is non-linear and space-variant. To resume :
- smoothing is formulated as a diffusive process,
- smoothing is performed at intra regions,
- smoothing is suppressed at region boundaries.

Mathematically, this is gotten by itteration as shown in the equation below :
$$
I_t = div\left(D\left(|\nabla I|\right)\nabla u\right)
$$ 
where $I_t$ is the time derivative of $I$, the image, $D$ is a diffusion function and $\nabla$ denotes the gradient. The diffusion function can be defined as in the next equations :
- $D(|\nabla I|) = e^{-\left(\frac{|\nabla I|}{k}\right)^2}$
- $D(|\nabla I|) = \frac{1}{1+\left(\frac{|\nabla I|}{k}\right)^2}$
- ...

The solution is found using an iterative approach on the discretized image grid (e.g. for 3D volume).
$$
I_{x,y,z}^{t+1} = I_{x,y,z}^{t} + \lambda \sum_{R=1}^{6} \left[ D\left(|\nabla _R I|\right) |\nabla _R I| \right]
$$
The gradient is then evaluated via finite differences.
$$
\nabla _1 I_{x,y,z} = I_{x-1,y,z} - I_{x,y,z}, \\
\nabla _2 I_{x,y,z} = I_{x+1,y,z} - I_{x,y,z}, \\
\nabla _3 I_{x,y,z} = I_{x,y-1,z} - I_{x,y,z}, \\
\nabla _4 I_{x,y,z} = I_{x,y+1,z} - I_{x,y,z}, \\
\nabla _5 I_{x,y,z} = I_{x,y,z-1} - I_{x,y,z}, \\
\nabla _6 I_{x,y,z} = I_{x,y,z+1} - I_{x,y,z}, \\
$$

The following example illustrates the application of the diffusion function $D(|\nabla I|) = e^{-\left(\frac{|\nabla I|}{k}\right)^2}$ and as it is shown, after a few iterations, the noise is removed while the heavy borders are conserved.

## Non-local image denoising

Unlike the local mean filter introduced before which take the mean value of the neighbourhood surrounding a target pixel, non-local means filtering takes the mean of all the pixels in the image, weighted by how similar these pixels are to the target pixel. This results in much greater post-filtering clarity, and less loss of detail in the image compared with local mean algorithms. Such an algorithm is given by the system below for an image $I$ and two pixels $p$ and $q$, $v(q)$ being de value of the pixel $q$ in the unfiltered image, $u(p)$ being the value of the pixel $p$ in the filtered image.
$$
\begin{cases}
    u(p) = \frac{1}{W(p)} \displaystyle\sum _{q \in I} v(q) \cdot w(p,q) \\
    W(p) = \displaystyle\sum _{q \in I} w(p,q)
\end{cases}
$$

$w(p,q)$ is the weighting fuction, that can be multiply defined. One common definition is the Gaussian weighting function $w(p,q) = e^{- \left(\frac{|B(q)^2 - B(p)^2|}{h^2}\right)}$. The Gaussian weighting function sets up a normal distribution with a mean $\mu =B(p)$ and a variable standard deviation $h$ in the neighbourhood of the pixel $p$. Thus :
- $B(p) = \frac{1}{|R(p)|} \displaystyle\sum _{i \in R(p)} v(i)$,
- $R(p) \subseteq I$ is a square region of pixels surrounding the pixel $p$, $|R(p)|$ being thus the number of pixels in the region $R(p)$.
